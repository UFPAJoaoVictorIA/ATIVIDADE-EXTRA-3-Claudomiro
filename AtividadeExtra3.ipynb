## Resolução do Projeto de Memória

**Parâmetros fornecidos:**
* Endereçamento do chip: 13 bits ($2^{13} = 8192$ posições).
* Capacidade total: 8 Gbits ($2^3 \times 2^{30}$ bits).
* Barramento de Dados (BD): 2 palavras por vez.
* Restrição: $BD \leq 64$ bits e $Palavra > Célula$.

```python
import math

# 1. Capacidade Total em bits
capacidade_total = 8 * (1024**3) # 8 Gbits

# 2. Endereços por chip (baseado nos 13 pinos)
enderecos_chip = 2**13

# 3. Barramento de Dados (BD) e Palavra
# Se BD <= 64 e BD = 2 * Palavra, então Palavra <= 32 bits.
# Vamos assumir o maior valor possível para eficiência: Palavra = 32 bits.
tamanho_palavra = 32
tamanho_bd = tamanho_palavra * 2

# 4. Célula
# A palavra deve ser maior que a célula. Assumiremos Célula = 8 bits (1 byte).
tamanho_celula = 8

# 5. Barramento de Endereços (BE) do sistema
# Total de células = Capacidade Total / Tamanho da Célula
total_celulas = capacidade_total / tamanho_celula
be_sistema = math.ceil(math.log2(total_celulas))

print(f"--- Especificações do Projeto ---")
print(f"Tamanho da Célula: {tamanho_celula} bits")
print(f"Tamanho da Palavra: {tamanho_palavra} bits")
print(f"Barramento de Dados (BD): {tamanho_bd} bits")
print(f"Barramento de Endereços (BE): {be_sistema} bits")
print(f"Capacidade Total: 8 Gbits")
```

### Descrição do Chip (Desenho Conceitual)

O sistema seria composto por um arranjo de chips.

**Cada Chip individual:**
- **Endereços:** 8192 (13 pinos: $A_0$ a $A_{12}$).
- **Saída de Dados:** Para formar uma palavra de 32 bits usando células de 8 bits, precisaríamos de 4 chips em paralelo por palavra.
- **Organização:** Como o BD lê 2 palavras (64 bits), o sistema ativa 8 chips simultaneamente para preencher o barramento de dados em cada ciclo de leitura.
